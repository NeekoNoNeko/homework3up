```python
import cv2

# 步骤1：原始图像处理，将彩色图像转换为灰色图像
image = cv2.imread('../dos/dataset4/manypeople.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 步骤2：加载级联分类器
faceCascade = cv2.CascadeClassifier('../dos/dataset4/haarcascade_frontalface_default.xml')

# 步骤3：人脸检测
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.04,
    minNeighbors=18,
    minSize=(8, 8)
)

# 步骤4：打印输出
print("发现{}张人脸！".format(len(faces)))
print("其位置分别是：")
print(faces)

# 步骤5：矩形框标注人脸并显示
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imshow("result", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
