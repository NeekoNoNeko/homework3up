```python
import cv2


# 步骤1：加载人脸分类器和微笑检测器
faceCascade = cv2.CascadeClassifier('../dos/dataset4/haarcascade_frontalface_default.xml')
smile = cv2.CascadeClassifier('../dos/dataset4/haarcascade_smile.xml')

# 步骤2：初始化摄像头
cap = cv2.VideoCapture(0,cv2.CAP_DSHOW)

while True:
    ret, image = cap.read()
    image = cv2.flip(image, 1)
    
    # if ret is None:
    #     break
    if not ret:
            break
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 人脸检测
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(5, 5)
    )

    # 步骤3：处理每张人脸
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # 提取人脸区域
        roi_color_face = image[y:y + h, x:x + w]
        roi_gray_face = gray[y:y + h, x:x + w]

        # 微笑检测（仅在人脸区域内）
        smiles = smile.detectMultiScale(
            roi_gray_face,
            scaleFactor=1.5,
            minNeighbors=25,
            minSize=(50, 50)
        )

        # 如果检测到微笑，显示文字
        # for (sx,sy,sw,sh)in smile:
        #     cv2.rectangle(roi_color_face,(sx,sy),(sx+sw,sy+sh),color=(0,255,255),thickness=2)
        #     cv2.putText(image, 'Smile', (x, y), cv2.FONT_HERSHEY_COMPLEX_SMALL,
        #                 1, (0, 255, 255), thickness= 2)
        
        for (sx,sy,sw,sh)in smiles:
            cv2.rectangle(roi_color_face,(sx,sy),(sx+sw,sy+sh),color=(0,255,255),thickness=2)
            cv2.putText(image, 'Smile', (x, y), cv2.FONT_HERSHEY_COMPLEX_SMALL,
                        1, (0, 255, 255), thickness= 2)


    # 显示结果
    cv2.imshow('dect', image)

    # 按 ESC 键退出
    if cv2.waitKey(25) == 27:
        break

# 释放资源
cap.release()
cv2.destroyAllWindows()
```
