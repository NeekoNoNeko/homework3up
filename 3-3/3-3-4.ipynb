```python
import cv2
import numpy as np

img = cv2.imread('..\\dos\\dataset3\\shape3.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img2 = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
c, h = cv2.findContours(img2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

arcl = cv2.arcLength(c[0], True)
ep = [0.1, 0.05, 0.01]
img3 = np.zeros(img.shape, np.uint8) + 255
cv2.drawContours(img3, c, -1, (0, 0, 255), 2)

for n in range(3):
    eps = ep[n] * arcl
    app = cv2.approxPolyDP(c[0], eps, True)
    img4 = img3.copy()
    cv2.drawContours(img4, [app], -1, (255, 0, 0), 2)
    cv2.imshow('appro %.2f' % ep[n], img4)
    cv2.waitKey(0)
```
