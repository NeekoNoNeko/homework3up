```python
import cv2
import numpy as np

img = cv2.imread('..\\dos\\dataset3\\shape2.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img2 = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
c, h = cv2.findContours(img2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for n in range(len(c)):
    length = cv2.arcLength(c[n], True)
    print('轮廓%s的长度：' % n, length)
```
