```python
import cv2
import numpy as np

img = cv2.imread('..\\dos\\dataset3\\shape4.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img2 = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
c, h = cv2.findContours(img2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rows, cols = img.shape[:2]
[vx, vy, x, y] = cv2.fitLine(c[0], cv2.DIST_L2, 0, 0.01, 0.01)
lefty = int((-x * vy / vx) + y)
righty = int(((cols - x) * vy / vx) + y)

img3 = np.zeros(img.shape, np.uint8) + 255
cv2.drawContours(img3, c, -1, (0, 0, 255), 2)
cv2.line(img3, (cols - 1, righty), (0, lefty), (255, 0, 0), 2)
cv2.imshow('FitLine', img3)
cv2.waitKey(0)
```
