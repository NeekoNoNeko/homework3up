```python
import cv2
import numpy as np

img = cv2.imread('..\\dos\\dataset3\\shape4.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img2 = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
c, h = cv2.findContours(img2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

x, y, w, h = cv2.boundingRect(c[0])
print('直边界矩形：', (x, y, w, h))

img3 = np.zeros(img.shape, np.uint8) + 255
cv2.drawContours(img3, c, -1, (0, 0, 255), 2)
cv2.rectangle(img3, (x, y), (x + w, y + h), (255, 0, 0), 2)
cv2.imshow('Rectangle', img3)
cv2.waitKey(0)
```
